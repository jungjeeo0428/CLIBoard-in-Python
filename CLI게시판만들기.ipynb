```python
# CLI게시판
print("CLI게시판에 오신걸 환영합니다")
board = []  #title, content 따로 리스트 하고 del삭제기능이 있다면  인덱스가 밀린다. 메인루프의 영향을 안받게 루프 밖에 있는 것이 중요하다
next_id = 1

while True:
  cmd = input("명령어:").strip()  #끝 공백은 봐주자
  if cmd == "exit":
    break
  elif cmd == "write article":
   line = input("제목:")
   line_text = input("내용:")

   article = {"id": next_id, #각자 리스트를 조립, 묶고싶다. list, dictionary, class
              "title": line, #keys values article 변수에 넣는다
              "content": line_text} #이건 조립이지 저장이 아니다.
   board.append(article) #저장을 잊지말자
   print("{}번글 생성".format(next_id))
   next_id += 1

  elif cmd == "article list":
    print("============================list==================================")
    print("번호 /        제목                    내용            ")
    for article in reversed(board):#보드에 있는만큼 다 뽑고 싶어서 for반복문이 필요하다 인덱스로 한다면 for(0,len,-1)도있다.
      print("번호{}/ 제목{}/ 내용{}".format(article["id"], article["title"],article["content"]))
    print("========================================================")


  elif cmd.startswith("article delete"):
    cmdBits = cmd.split(" ")
    int(cmdBits[2]) == next_id

    if int(cmdBits[2]) in board:
      board.remove(board[next_id])
    else:
      print(str(cmdBits[2])+"번글은 존재하지 않습니다.")



  else:
    print("지원하지 않는 기능입니다")



print("==CLI 게시판 종료 ==")
```
