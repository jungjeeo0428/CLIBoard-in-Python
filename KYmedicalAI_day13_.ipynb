```python
# CLI게시판


print("CLI게시판에 오신걸 환영합니다")
board = []
next_id = 1
from datetime import datetime



while True:
  cmd = input("명령어:").strip()
  if cmd == "exit":
    break
  elif cmd == "write article":
   line = input("제목:")
   line_text = input("내용:")

   regDate = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
   updateDate = regDate
   article = {"id": next_id,
              "regDate": regDate, "updateDate" : updateDate,
              "title": line,
              "content": line_text}
   board.append(article)

   print("{}번글 생성".format(next_id))
   next_id += 1

  elif cmd == "article list":
    today = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print("==========================list==========================",
          "\n번호 /   시간/     제목           내용            ")
    for article in reversed(board):

      if today.split(" ")[0] != article["regDate"].split(" ")[0]:
            print("   {}      /     {}       /     {}          /     {}   ".format(article["id"],
                                                                                   article["title"],
                                                                                   article["content"],
                                                                                   article["regDate"].split(" ")[0]))
      if today.split(" ")[0] == article["regDate"].split(" ")[0]:
            print("   {}      /     {}       /     {}          /     {}   ".format(article["id"],
                                                                                   article["title"],
                                                                                   article["content"],
                                                                                   article["regDate"].split(" ")[1]))
      print("=========================================")


  elif cmd.startswith("article delete"):
    cmdBits = cmd.split(" ")
    target = int(cmdBits[2])

    for article in board:
      if article["id"] == target: #우리가 article delete 3이라고 쳤을 때 3이 id가 되어 이미 board 딕셔너리에  3번 방에 입장한 것이다.
        board.remove(article)
        print("{}번 글이 삭제되었습니다.".format(target),
              "\n--------------------------")
        break
    else:
      print((cmdBits[2])+"번글은 존재하지 않습니다.")

  elif cmd.startswith("article edit"):
    cmdBits = cmd.split(" ")
    target == int(cmdBits[2])
    found = False

    for article in board:

      if article["id"] == target:

       print("기존 제목: {}".format(article["title"]),
             "\n기존 내용: {}".format(article["content"]))

       article["title"] = input("새 제목:")
       article["content"] = input("새 내용:")
       article["updateDate"] = datetime.now().strftime("%Y-%m-%d %H:%M")
       found = True

       print("제목과 내용이 수정되었습니다.")
       break
      if found == False:
        print((cmdBits[2])+"번글은 존재하지 않습니다.")

  elif cmd.startswith("article detail"):
       cmdBits = cmd.split(" ")
       detailid = int(cmdBits[2])
       found = False
       if len(cmdBits) < 3:
         print("상세보기할 글 번호를 입력해주세요.")
         continue

       for article in board:
          if article["id"] == detailid:
           print("번호 : {}".format(article["id"]),
           "\n날짜 : {}".format(article["regDate"]),
           "\n수정 날짜 : {}".format(article["updateDate"]),
           "\n제목 : {}".format(article["title"]),
           "\n내용 : {}".format(article["content"]))
           found = True
           break
       if found == False:
          print("{}번 글은 없습니다".format(cmdBits[2]))

  else:
    print("지원하지 않는 기능입니다")



print("==CLI 게시판 종료 ==")
```
